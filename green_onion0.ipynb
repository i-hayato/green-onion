## CSVファイルを一つにまとめる

```python
import pandas as pd
import glob
import os

# 対象CSVのパスを取得
csv_files = glob.glob("data/tsuji/*.csv")

# 各CSVを読み込んでDataFrameのリストに
df_list = []
for file in csv_files:
    df = pd.read_csv(file, encoding="cp932", parse_dates=["日付"])
    df["ファイル名"] = os.path.basename(file)  # 元ファイル名を追加しておくと便利
    df_list.append(df)

# すべて結合
df_all = pd.concat(df_list, ignore_index=True)
```

```python
# 日付、時刻の変換や不要行削除
df_all["日付"] = pd.to_datetime(df_all["日付"], format="%Y/%m/%d", errors="coerce")
df_all["時刻"] = df_all["時刻"].str.replace("*", "", regex=False)
df_all["時刻"] = pd.to_datetime(df_all["時刻"], format="%H:%M", errors="coerce").dt.time

# 全NaN列の削除
df_all = df_all.dropna(axis=1, how="all")

# すべてが0の列の削除
df_all = df_all.loc[:, ~(df_all == 0).all()]
```

```python
print(df_all.shape)
print(df_all.dtypes)
df_all.head()
```

```python
import pandas as pd

df = pd.read_csv(
    "data/tsuji/p0005211_環境データ_12号棟_202408.csv",
    encoding="cp932",
    parse_dates=["日付"]
)

df["日付"] = pd.to_datetime(df["日付"], format="%Y/%m/%d", errors="coerce")
df["時刻"] = df["時刻"].str.replace("*", "", regex=False)
df["時刻"] = pd.to_datetime(df["時刻"], format="%H:%M", errors="coerce").dt.time

print(df.dtypes)
df.head()
```
